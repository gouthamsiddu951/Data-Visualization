## Data visualization

### Subtask:
Generate bar charts to visualize runs scored by players and scatter plots to analyze the relationship between strike rate and runs.


**Reasoning**:
Generate a bar chart for runs by players and a scatter plot for strike rate vs. runs, adding titles and labels to both for clarity.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar chart of Runs by Player
plt.figure(figsize=(15, 7))
sns.barplot(x='Player', y='Runs', data=df.head(20)) # Display top 20 players for readability
plt.xticks(rotation=90)
plt.title('Runs Scored by Players (Top 20)')
plt.xlabel('Player')
plt.ylabel('Runs')
plt.tight_layout()
plt.show()

# Create a scatter plot of SR vs Runs
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SR', y='Runs', data=df)
plt.title('Strike Rate vs. Runs')
plt.xlabel('Strike Rate (SR)')
plt.ylabel('Runs')
plt.show()
```
